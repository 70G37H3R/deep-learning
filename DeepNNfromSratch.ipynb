# 1. Load the data

```python
import pandas as pd

train_data = pd.read_csv('https://raw.githubusercontent.com/70G37H3R/house_prices/main/train.csv')
test_data = pd.read_csv('https://raw.githubusercontent.com/70G37H3R/house_prices/main/test.csv')
```

```python
train_data.head(10)
```

```python
test_data.head(10)
```

# 2. Get Features

```python
x_train = train_data.iloc[:, 1:-1]
x_train.head(10)
```

```python
x_test = test_data.iloc[:, 1:]
x_test.head(10)
```

# 3. Get Labels

```python
y_train = train_data.iloc[:,-1:]
y_train.head(10)
```

# 4. Preprocessing data

## 4.1 Drop NULL columns

```python
x_train.isna().sum().sort_values(ascending = False).head(20)
```

```python
x_train = x_train.drop(labels=['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 'LotFrontage'], axis=1)
x_train.head(20)
```

```python
x_test = x_test.drop(labels=['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 'LotFrontage'], axis=1)
x_test.head(20)
```

## 4.2 Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder

# Select non-numerical categorical columns
non_num_cols = x_train.select_dtypes(exclude=["int64", "float64"]).columns.tolist()
encoder = OrdinalEncoder()

# Fit the encoder on the selected columns
encoder.fit(x_train[non_num_cols])

# Transform the selected columns
encoded_cols = encoder.transform(x_train[non_num_cols])

# Replace the original columns with the encoded ones
x_train[non_num_cols] = encoded_cols

x_train.head(20)
```

```python
# Fit the encoder on the selected columns
encoder.fit(x_test[non_num_cols])

# Transform the selected columns
encoded_cols = encoder.transform(x_test[non_num_cols])

# Replace the original columns with the encoded ones
x_test[non_num_cols] = encoded_cols

x_test.head(10)
```

## 4.3 Normalization - MinMaxScaler

```python
x_train = (x_train-x_train.min())/(x_train.max()-x_train.min())
x_train.head(10)
```

```python
x_test = (x_test-x_test.min())/(x_test.max()-x_test.min())
x_test.head(5)
```

## 4.4 Handle Missing Values

```python
x_train.fillna(x_train.mean(), inplace=True)
x_train.fillna(0, inplace = True)
x_train.head(20)
```

```python
x_test.fillna(x_test.mean(), inplace=True)
x_test.fillna(0, inplace = True)
x_test.head(20)
```

```python
x_train.isna().sum().sort_values(ascending = False).head(20)
```

```python
x_test.isna().sum().sort_values(ascending = False).head(20)
```

# 5. Features Selection

```python
from sklearn.linear_model import LinearRegression
from mlxtend.feature_selection import SequentialFeatureSelector as SFS

# Initialize linear regression model
lr = LinearRegression()

# Backward Elimination
sfs_backward = SFS(lr,
                   k_features=(1, len(x_train.columns)),
                   forward=False,
                   floating=False,
                   scoring='r2',
                   cv=5)

# Fit the backward selection models
sfs_backward = sfs_backward.fit(x_train, y_train)

# Get the selected features
selected_features_backward = list(sfs_backward.k_feature_names_)

print("Selected features (Backward Elimination):", selected_features_backward)
```

```python
x_train_backward = x_train[selected_features_backward]
x_train_backward.head(5)
```

```python
x_test_backward = x_test[selected_features_backward]
x_test_backward.head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Concatenate x_train and y_train into a single DataFrame
data = pd.concat([x_train_backward, y_train], axis=1)
columns = data.columns

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Plotting the correlation matrix as a heatmap
plt.figure(figsize=(25, 25))
sns.set(font_scale=1.25)
sns.heatmap(correlation_matrix, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=columns, xticklabels=columns, cmap="coolwarm")
plt.show()
```

# 6. Training

## 6.1 Build NN architecture

```python
import tensorflow as tf
import numpy as np

class DenseLayer(tf.Module):
    def __init__(self, input_size, output_size, activation='relu'):
        super(DenseLayer, self).__init__()
        self.weights = tf.Variable(tf.random.normal([input_size, output_size], dtype=tf.float32))
        self.bias = tf.Variable(tf.zeros([output_size], dtype=tf.float32))
        self.activation = activation

    def __call__(self, inputs):
        outputs = tf.matmul(inputs, self.weights) + self.bias
        if self.activation == 'relu':
            outputs = tf.nn.relu(outputs)
        elif self.activation == 'linear':
            pass  # No activation function needed for linear activation
        else:
            raise ValueError("Activation function not supported.")
        return outputs

class NeuralNetwork(tf.Module):
    def __init__(self, input_size, output_size):
        super(NeuralNetwork, self).__init__()
        self.dense_layers = []
        self.input_size = input_size
        self.output_size = output_size

    def add_dense_layer(self, units, activation='relu'):
        layer = DenseLayer(self.input_size, units, activation)
        self.input_size = units
        self.dense_layers.append(layer)

    def __call__(self, inputs):
        x = inputs
        for layer in self.dense_layers:
            x = layer(x)
        return x
```

```python
# Define hyperparameters
input_size = 24  # Number of features
output_size = 1  # Output size (predicted house price)

# Instantiate the model
NN_model = NeuralNetwork(input_size, output_size)
```

```python
# Add layers to the model
NN_model.add_dense_layer(128, activation='relu')
NN_model.add_dense_layer(256, activation='relu')
NN_model.add_dense_layer(256, activation='relu')
NN_model.add_dense_layer(256, activation='relu')
NN_model.add_dense_layer(1, activation='linear')
```

## 6.2 Build Training Loop

```python
# Training loop
learning_rate = 0.001
num_epochs = 500
optimizer = tf.optimizers.Adam(learning_rate)
```

```python
# Convert input and output data to float32
x_train_final = tf.cast(x_train_backward, dtype=tf.float32)
y_train_final = tf.cast(y_train, dtype=tf.float32)
```

```python
# Initialize variables for early stopping
best_loss = np.inf  # Initialize the best loss to infinity
patience = 10  # Number of epochs to wait for improvement
count = 0  # Counter to track the number of epochs without improvement
```

```python
for epoch in range(num_epochs):
    with tf.GradientTape() as tape:
        predictions = NN_model(x_train_final)
        loss = tf.reduce_mean(tf.abs(predictions - y_train_final))

    gradients = tape.gradient(loss, NN_model.trainable_variables)
    optimizer.apply_gradients(zip(gradients, NN_model.trainable_variables))

    # Validation
    val_predictions = NN_model(x_train_final)
    val_loss = tf.reduce_mean(tf.abs(val_predictions - y_train_final))

    print(f'Epoch {epoch + 1}, Training Loss: {loss.numpy()}, Validation Loss: {val_loss.numpy()}')

    # Check if validation loss has improved
    if val_loss < best_loss:
        best_loss = val_loss
        count = 0
    else:
        count += 1

    # Check if we should stop early
    if count >= patience:
        print("Early stopping! No improvement for {} epochs.".format(patience))
        break
```
